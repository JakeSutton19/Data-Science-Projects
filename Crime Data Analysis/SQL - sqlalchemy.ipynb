```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
file = "Census_Data.csv"
output = 'output.csv'
```

```python
engine = create_engine('sqlite://', echo=False)
df = pd.read_csv(file)

df.to_sql('census', engine, if_exists='replace', index=False)

results = engine.execute('Select * from census')

final=pd.DataFrame(results, columns=df.columns)
final.to_csv(output, index=False)
```
