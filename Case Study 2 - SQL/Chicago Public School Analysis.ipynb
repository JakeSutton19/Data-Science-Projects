```python
import sqlite3
import pandas as pd
from sqlalchemy import create_engine
```

```python
census_file = "Census_Data.csv"
crime_file = "Chicago_Crime_Data-v2.csv"
school_file = "Chicago_Public_Schools_-_Progress_Report_Cards__2011-2012-v3.csv"
output = 'output.csv'
```

```python
engine = create_engine('sqlite://', echo=False)
```

```python
df = pd.read_csv(crime_file)
df.to_sql('crime', engine, if_exists='replace', index=False)
sql_action = engine.execute('Select * from crime')

result=pd.DataFrame(sql_action, columns=df.columns)
result.head()
```

Problem 1: Find the total number of crimes recorded in the Crime table

Answer

```python
sql_action = engine.execute('Select COUNT(*) as "Crimes Count" from crime')

result=pd.DataFrame(sql_action)
result
```

Problem 2: Retrieve first 5 rows from Chicago_Crime_Data

Answer:

```python
sql_action = engine.execute('Select * from crime LIMIT 5')

result=pd.DataFrame(sql_action, columns=df.columns)
result.head()
```

Problem 3: How mamy crimes involve an arrest?

Answer:

```python
sql_action = engine.execute('Select COUNT(ARREST) as "Arrests Count" from crime \
                            where ARREST > 0')

result=pd.DataFrame(sql_action)
result
```

Problem 4: Which unique types of crimes have been recorded on Street locations?

Answer:


```python
sql_action = engine.execute('Select PRIMARY_TYPE as "Types of Crimes on Streets" from crime \
                            where LOCATION_DESCRIPTION = "STREET" \
                            Group By PRIMARY_TYPE')

result=pd.DataFrame(sql_action)
result
```

Problem 5: In the Census Data list all community areas whose names start with the letter 'O'

Answer:

```python
df = pd.read_csv(census_file)
df.to_sql('census', engine, if_exists='replace', index=False)
sql_action = engine.execute('Select * from census')

result=pd.DataFrame(sql_action, columns=df.columns)
result.head()
```

```python
sql_action = engine.execute('Select COMMUNITY_AREA_NAME as "Community Area Names that start with O" from census \
                            where COMMUNITY_AREA_NAME like "O%"')

result=pd.DataFrame(sql_action)
result
```

Problem 6: Which schools in community areas 10 to 35 and have a harship index above 20? 

Answer:


```python
df = pd.read_csv(school_file)
df.to_sql('schools', engine, if_exists='replace', index=False)
sql_action = engine.execute('Select * from schools')

result=pd.DataFrame(sql_action, columns=df.columns)
result.head(5)
```

```python
sql_action = engine.execute('Select COMMUNITY_AREA_NUMBER, NAME_OF_SCHOOL from schools \
                            join census as new_CD on new_CD.COMMUNITY_AREA_NUMBER = schools.COMMUNITY_AREA_NUMBER \
                            where new_CD.COMMUNITY_AREA_NAME between 10 and 35 \
                            and new_CD.HARDSHIP_INDEX > 20')

result=pd.DataFrame(sql_action)
result
```
